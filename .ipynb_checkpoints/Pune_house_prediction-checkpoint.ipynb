```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer, KNNImputer
```

```python
df=pd.read_csv("Pune_House_Data.csv")
```

```python
df.drop(columns=['area_type','availability','society'],inplace=True) 
```

```python
df.describe()
```

```python
df['total_sqft']
```

```python

def clean_arr(x):
    try:
        if '-' in str(x):
            a, b = x.split('-')
            return (float(a) + float(b)) / 2
        return float(x)
    except:
        return np.nan

df['total_sqft'] = df['total_sqft'].apply(clean_arr)
```

```python
# Calculate max price per location and find the top 10
max_prices = df.groupby('site_location')['price'].max().sort_values(ascending=False).head(10)

# Plot 1: Top 10 Locations by Max Price
plt.figure(figsize=(12, 8))
max_prices.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Top 10 Locations by Maximum Price')
plt.xlabel('Location')
plt.ylabel('Maximum Price (in Lakhs )')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_locations_max_price.png')
plt.show()

# Calculate mean price per location and find the top 10
mean_prices = df.groupby('site_location')['price'].mean().sort_values(ascending=False).head(10)

# Plot 2: Top 10 Locations by Mean Price
plt.figure(figsize=(12, 8))
mean_prices.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Top 10 Locations by Mean Price')
plt.xlabel('Location')
plt.ylabel('Mean Price (in Lakhs )')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_10_locations_mean_price.png')
plt.show()

print("Plots saved as top_10_locations_max_price.png and top_10_locations_mean_price.png")
```
